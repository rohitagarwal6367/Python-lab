```python
limit = 2000000
sum_primes = 0

for num in range(2, limit):
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        sum_primes += num

print("Sum of primes below two million:", sum_primes)
```

```python
a, b = 1, 2
sum_even = 0

while b <= 4000000:
    if b % 2 == 0:
        sum_even += b
    a, b = b, a + b

print("Sum of even Fibonacci numbers:", sum_even)
```

```python

#Checking Whether a Number is Positive, Negative, or Zero
num = float(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
# Checking Divisibility of a Number by 5 and 11
num = int(input("Enter a number: "))

if num % 5 == 0 and num % 11 == 0:
    print("Divisible by both 5 and 11")
else:
    print("Not divisible by both 5 and 11")
```
